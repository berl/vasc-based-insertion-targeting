## Validating radii 

```python
import SWC
# import an SWC file and allow it to ground its pixel coordinates in a real measurement
mpp = .92379 # from the 2p Z-stack metadata
pre000_swc = SWC.SWC(path='../../vascdata/pre000/pre000-8bitGF5552-wrad7-distancemask.swc', microns_perpixel=mpp)
len(pre000_swc) # the number of segments represented in this swc
```

```python
reduce(lambda x, y: len(y) + x, pre000_swc, 0) # the number of pieces represented in this swc 
```

Let's get a list of every radii from every segment in this SWC.

```python
pre000_rads = reduce(lambda x, y: np.concatenate((x, y.micron_rads)), pre000_swc, np.array([]))
len(pre000_rads)
```

Now let's look at the distribution of our radii.

```python
n, bins, patches = plt.hist(pre000_rads, bins=20)
plt.plot(bins)
plt.xlabel('radius')
plt.ylabel('number of vessels')
plt.title('Distribution of radii for pre000')
```

```python
# find the mean, standard deviation, median, and range
print 'median: '+str(np.median(pre000_rads))
print 'mean  : '+str(np.mean(pre000_rads))
print 'std   : '+str(np.std(pre000_rads))
print 'min   : '+str(min(pre000_rads))
print 'max   : '+str(max(pre000_rads))
```

We can see from the histogram that the distribution is largely unimodal with a long tail to the right. Due to the long tail, median provides a better measure of central tendency in this case. We do not have the ability to detect vessels with a radius smaller than the .92379, as that is also the limit of our resolution. This distribution is similar to that which was found by Blinder (2013, Fig 3e). Our population statistics are also comparable, though not identical -- their mean/median were both "near two." Blinder's methods (use of a fluorescent gel) allowed them to detect smaller vessels than us, which can explain our larger detected mean/median. Further, Blinder and Tsai (2009), which reports an "average" radius of 1.75 - 2$\mu m$, both report the average radius of microsvessels only and not both microvessels and larger penetrating vessels. While a measure like median should be insensitive to the relatively few contributions of penetrating vessels, our inclusion of them in our summary statistics may have further contributed to the inflation of our mean. 

Next, we look at the function of radius over depth, to compare with Blinder (2013, Fig 3b). To do that, we will first organize our SWC as a function of z-coordinate, and then plot the medians at each level. We expect a distribution that remains constant as a function of depth, which is what we get. This is consistent with the result in Blinder. 

```python
import numpy.ma as ma # we have to mask out nan values

stack_depth = 512
layers = np.zeros((stack_depth), dtype=np.dtype('object'))
layers[:] = np.nan # so we don't plot zeros
for segment in pre000_swc:
    for piece in segment:
        if piece.z < stack_depth:
            if np.all(layers[piece.z] == np.nan):
                layers[piece.z] = np.array([piece.rad])
            else:
                layers[piece.z] = np.append(layers[piece.z], piece.rad)
layer_median = map(lambda x: np.median(x) * mpp, layers)
masked_layer_median = ma.masked_where(np.isnan(layer_median), layer_median)
plt.figure(figsize=(4,8))
plt.plot(layer_median, np.arange(512*mpp, step=mpp))
plt.gca().invert_yaxis()
plt.xlabel('median radius ($\mu m$)')
plt.ylabel('depth ($\mu m$)')
plt.title('radius as a function of depth')
```

Finally, we measure total detected vascular volume and compare it to that reported in Tsai (2009). We do this by creating a binary mask based on the SWC file generated, with radius measurements, by vaa3d and computing the volume occupied by vessels.

```python
import tiff.tifffile as tiff
binary_mask = tiff.imread('../../vascdata/pre000/pre000-8bitGF5552-maskwrad7-cropped.tif')
bm_max = binary_mask.max()
binary_mask[binary_mask == bm_max] = 1 # ensure max value of binary mask is one
print 'mask maximum : '+str(binary_mask.max())
print 'shape of the mask : '+str(binary_mask.shape)
print 'number of pixels in the mask : '+str(binary_mask.size)
print 'number of non-zero pixels in mask : '+str(binary_mask.sum())
print 'percent volume : '+str(binary_mask.sum() * 100 / binary_mask.size) +'%'
```

Tsai reports a vascular density of ~1%, but notes that their estimate is lower by a factor of two to three than other estimates in the literature -- our estimate of density is in line with the other estimates cited. They explain this difference as follows: 

>With regard to image analysis, we attribute our lower measure of vascular volume to the use of local thresholds, which accounts for the point spread function of the optical system, to accurately segment the vascular images ... The common practice of applying a threshold whose value is close to the noise floor can result in artificially enlarged diameters...

As we do not use local thresholds, it makes sense that our measure of vascular density, and of radii, will be inflated compared to Tsai. 

#### Comparison with line profiles

```python
import compare
```

After checking visually and comparing aggregate statistics to Blinder (2013), we decided to compare the Vaa3d radius fill to radii produced by a simpler line profile approach -- which is both easy to validate visually and easy to compare to the Vaa3d data. Instead of maximizing the radius by trying a variety of angles for the profile, we used a single profile that was orthogonal to the direction of travel of the segment in the xy plane as given by the vector $p_d$ s.t. $p_{nd} = p_{n+1} - p_{n-1}$ where $p_n$ is the current piece. 

```python
compare.filled_radius_line_profile(pre000_swc, '../../vascdata/pre000/pre000-8bitGF5552.tif')
# note the summary statistics are slightly different: this is because we excluded segments 
# close to the borders of the image to ease the application of the line profile
```

##### Failure

The radius results appear internally consistent, but not perfect. In many cases, it appears that vaa3d is reporting a larger radius than the line profile. To investigate whether or not this represented an error on the part of vaa3d, we looked closely at several of these cases. To do so, we displayed randomly selected examples from the subset of cases when $r_{vaa3d} - r_{profile} > 6$.

```python
reload(compare)
compare.filled_radius_line_profile(pre000_swc, '../../vascdata/pre000/pre000-8bitGF5552.tif', random_display=True, main_display=False,
                                   lessthan=False)
```

The principal cause of the disagreement between vaa3d and the line profiles is on large vessels. A known issue with the current tracing software is a failure to correctly identify the centerline of large vessels and, instead, following minor gradients within the body of the vessel. This snaking behavior causes our orthogonal vector to often not be orthogonal to the actual vessel, which invites the line profile to detect minor gradients in the body of the vessel, not the vessel boundaries. 

Vaa3d also outperforms the line profiles in cases with uneven luminance, the line profile will find a small peak, while vaa3d will detect the boundary of the luminance with restpect to the background. 

The line profile may outperform vaa3d in cases in which there is a thin, high luminance vessel surrounded by noise luminance. Vaa3d will include the surrounding noise, thus overestimating the vessel radius while the line profile will detect only the strong peak of thin vessel. These cases, however, are not common. 

In general, we are satisfied that vaa3d outperforms a simple line profile approach to judging vessel radius.

## Length

```python
# brief note on length, Tsai computes a length of .88m, and cites a .7 - 1.2m range of vasculature in 1mm^3, 
# figured I'd check against that now
length_list = map(lambda x: x.length(), pre000_swc) # x.length() returns in microns
print 'mean length : '+str(np.mean(length_list))
print 'median length : '+str(np.median(length_list))
print 'std : '+str(np.std(length_list))
print 'range : '+str(min(length_list))+','+str(max(length_list))
total_length = sum(length_list) / 1000000 # convert to meters
print 'total length : '+str(total_length)+'m'
# the binary mask is the same size as our data, so we'll use that
volume_seen = binary_mask.size * mpp # in microns
volume_mm3 = 1000 ** 3 # in microns
seen = volume_seen / volume_mm3
len_per_mm3 = total_length / seen
print 'length per mm^3 : '+str(len_per_mm3)+'m'
```

We are just below the observed range for manual tracing cited by Tsai, which is consistent with some automatic segmentation failures and some failures to detect small vessels. 

The above population statistics on length are meaningless with regards to vessel length.
