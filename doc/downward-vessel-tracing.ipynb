#### Distance Binary Trace

Documentation of penetrating vessel tracing for the re-snake trace of a distance transform of a binary mask from an original snake tracing. 

This data has been:
1. Converted to 8bit
2. Gaussian filtered, window-size was 5x5x5 with $\sigma$=2
3. Snake traced by Vaa3d
4. Radius filled using the snake trace swc and the 8bit, gaussian filtered image
5. The radius filled snake trace swc was converted to a binary mask (cylindrical) using the neuron utilities->swc_to_maskimage_cylinder_unit plugin
6. The binary mask from (5) was grayscale image distance transformed with a threshold of .5
7. The distance transformed image from (6) was snake traced
8. The snake trace swc from (7) was radius filled using the 8bit, gaussian filtered image from (2)

## Comparison of methods

```python
methods = {'basic trace':basic_trace, 'distance binary trace':db_trace}
for x in ident_penetrators:
    print '-------- '+x+' --------'
    for t in methods.keys():
        print '---- '+t+' ----'
        methods[t][x].inspect()
```

```python
basic_super = SWC.SuperSegment([basic_trace[x] for x in ident_penetrators], ident='basic super')
db_super = SWC.SuperSegment([db_trace[x] for x in ident_penetrators], ident='distance binary super')

print '----------'
a = db_super.inspect()
print '----------'
a = basic_super.inspect()
```

From these aggregate analyses, we note that the distance binary method records greater downwardness and larger radii than the basic pre-processing method and records a much smaller amount of length. 

```python
from mpl_toolkits.mplot3d import Axes3D

max_x = 0; max_y = 0
for name in ident_penetrators:
    for meth in methods.keys():
        coords = methods[meth][name].xyzs
        max_x = max(max(coords[:, 0]) - min(coords[:, 0]), max_x)
        max_y = max(max(coords[:, 1]) - min(coords[:, 1]), max_y)

fig = plt.figure(figsize=(4,12))
for i, name in enumerate(ident_penetrators):
    for j, t in enumerate(methods.keys()):
        v_axe = fig.add_subplot(4, 2, i*2 + j + 1, projection='3d')
        v_axe.set_title(name)
        seg = methods[t][name]
        coords = seg.xyzs
        v_axe.plot(coords[:, 0], coords[:, 1], coords[:, 2], ['bo','ro'][j], label=t, alpha=.75)
        v_axe.legend(fontsize=6, loc='lower right')
        interv_x = v_axe.get_xaxis().get_view_interval()
        mid_int_x = interv_x[0] + interv_x[1]/2
        interv_y = v_axe.get_yaxis().get_view_interval()
        mid_int_y = interv_y[0] + interv_y[1]/2
        v_axe.set_xlim(mid_int_x - max_x/2, mid_int_x + max_x/2 )
        v_axe.set_ylim(mid_int_y + max_y/2, mid_int_y - max_y/2)
        v_axe.set_zlim(coords[:,2].max(), 0)
        
```
